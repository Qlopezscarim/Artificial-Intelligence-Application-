**In the discussion of activation functions**<br>
RELU: standard and typically used <br>
Sigmoid : used for output if it all (between 0 and 1, thus good for a final output) <br>
Tanh : Universially better than sigmoid, occasionally used over REL, centered at zero <br>
Leaky ReLU : arguably better then RELU as it's slope is not zero while negative for gradients
